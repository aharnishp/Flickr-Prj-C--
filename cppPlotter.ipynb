```python
import matplotlib.pyplot as plt
```

```python
# read file into a list
with open('music/outputs_inter_over_Gf.csv', 'r') as f:
    dataGf = f.readlines()


# remove whitespace characters like `\n` at the end of each line
dataGf = [x.strip() for x in dataGf]

# remove the last element of the list
dataGf.pop()

# convert the list to a float list
dataGf = [float(i) for i in dataGf]

# reverse the list
dataGf.reverse()

plt.plot(dataGf)
```
